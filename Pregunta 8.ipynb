```python
def Spline(x: float, x0: float, pars: dict[str, float]) -> float:
    a = pars["a"]
    b = pars["b"]
    c = pars["c"]
    d = pars["d"]
    return a + b * (x - x0) + c * (x - x0) ** 2 + d * (x - x0) ** 3
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
s = [
    {"a": 1, "b": 5.5, "c": 0, "d": -1.5},
    {"a": 5, "b": 1, "c": -4.5, "d": 1.5},
]
for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 20)
    _y = Spline(_x, x_i, s[i])
    plt.plot(_x, _y, color="red")

plt.scatter(xs, ys)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.show()
```

# Grafique los splines cúbicos tal que cumplan con una pendiente de m=-3 en los puntos dados.

```python
import matplotlib.pyplot as plt
import numpy as np

xs = [-1, 0, 1]
ys = [1, 5, 3]
m = -3 

d1 = 1
s_ajustado = [
    {
        "a": 1,
        "b": 19 - d1,
        "c": 2 * d1 - 23,
        "d": 8 - d1,
    },
    {
        "a": 5,
        "b": -3,
        "c": 1 - d1,
        "d": d1,
    },
]

for i, x_i in enumerate(xs[:-1]):
    _x = np.linspace(x_i, xs[i + 1], 100)
    _y = Spline(_x, x_i, s_ajustado[i])
    plt.plot(_x, _y, color="red", label=f"Spline {i}")

plt.scatter(xs, ys, color="blue")
x_line = np.linspace(-0.5, 0.5, 10)
y_line = m * x_line + ys[1]  # y = m(x - x1) + y1 con x1 = 0
plt.plot(x_line, y_line, '--', label='Pendiente modificada (m = -3)', color="blue")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Interpolación con splines cúbicos")
plt.legend()
plt.grid(True)
plt.show()
```
